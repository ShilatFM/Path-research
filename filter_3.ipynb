```python
df = pd.read_pickle("data/to_pickle.pk1.xz")
```

```python
def area(filtering_table, top_left, bottom_right):
    points_in_square = filtering_table[(filtering_table.x.between(top_left[0], bottom_right[0])) &
                          (filtering_table.y.between(bottom_right[1], top_left[1]))] #all the point in the square
    return points_in_square
```

```python
def draw(table):
    table = table.groupby(['filename', 'obj']).size()
    df_by_obj = df.set_index(['filename', 'obj']).sort_index()
    im = imread('paths0.png')
    imshow(im)
    for t in table.index:
        s_o = df_by_obj.loc[t]
        plot(s_o.x, s_o.y, label = t[1])
    legend(loc=9, bbox_to_anchor=(1.1, 1))
```
